```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

import google.colab.drive
google.colab.drive.mount('gdrive')
DATA_PATH = "/content/gdrive/My Drive/40_スマートエスイー/k13_6_異常検知/AirPassengers.csv"
data = pd.read_csv(DATA_PATH)
#data = pd.read_csv("AirPassengers.csv")
df = pd.Series(data['#Passengers']) 
df.index = pd.to_datetime(data['Month']) 
df
```

```python
train_data = df[0:-12]
test_data = df[-12:-1]
print(train_data)
print(test_data)
```

```python
dec = sm.tsa.seasonal_decompose(train_data)
```

```python
plt.plot(dec.seasonal)
plt.show()
```

```python
plt.plot(dec.trend)
plt.show()
```

```python
plt.plot(dec.resid)
plt.show()
```

```python
sarima = sm.tsa.SARIMAX(train_data, order=(3, 1, 2), seasonal_order=(1, 1, 1, 12), enforce_stationarity = False, enforce_invertibility = False).fit()

print(sarima.aic)
```

```python
#
#pred = sarima.predict(test_data.index[0], test_data.index[-1])
pred = sarima.predict('1960-01-01','1960-12-01')
```

```python
print(pred)
```

```python
plt.plot(df, "blue")
plt.plot(pred, "red")
plt.show()
```
